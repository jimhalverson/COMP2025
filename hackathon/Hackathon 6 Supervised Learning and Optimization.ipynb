# Title: Exoplanets

*Abstract:* In this problem, we will use supervised learning to predict whether an exoplanet is habitable or not. We will use the [NASA Exoplanet Archive](https://exoplanetarchive.ipac.caltech.edu/) to get data on exoplanets. Then look up how to compute whether or not a planet is in the habitable zone of its star, and use that as the label for your supervised learning problem. You can use any supervised learning algorithm you like, but I suggest starting with logistic regression and then a feedforward neural network with binary cross-entropy loss. Note: this is a *binary* classification problem, you're just trying to figure out from the data whether a given planet is habitable.

If this is too easy and you finish early, use the data to analyze the location of the planets in the galaxy and estimate the number of exoplanets in the universe. Compare your naive estimate to expectations in the literature.

# Title: Ising Phase Transitions with CNN

(Note: this is one of of the homework problems!)

*Abstract:*
Can neural networks detect phase transitions? We'll test this idea in this problem, using a well-studied case where we understand everything. In class we solved the 2d Ising model and saw a phase transitions at $T_c\sim 2.6K$. Below this critical temperature we were in the ordered phase, and above it we were in the disordered phase.

In this problem we would like to learn to distinguish Ising samples from well below or above the critical temperature. To do so:
1. Use the Monte Carlo algorithm *from class* to generate $10,000$ samples at both $T=0.5K$ and $T=5.5K$. Label these samples $0$ for ordered and $1$ for disordered.
2. With $90/10$ train-test split, use a NN (probably CNN is best, but feel free to experiment) to classify the samples as ordered or disordered, plotting both train and validation (test) accuracy as a function of training epoch.
3. We want to test whether the **already trained** NN can detect the critical temperature. To do so, generate $10,000$ samples for all $T$ in $[1.0,1.5,2.0,2.5,3.0,3.5.,4.0,4.5,5.0]$ and plot the average network prediction and error bars on these samples as a function of $T$. Does it see the phase transition?

## Title: Neural Quantum States or Physics Informed Neural Networks

*Abstract:*  In this problem, write a general code for solving the Schrodinger equation, subject to boundary or initial conditions, with a neural network. The essential idea is that the neural network is $\Psi(x)$, and you're trying to regress on its parameters to solve the Schrodinger equation. You might also include the energy into the loss to try to specifically force it to the ground state. Given the general structure, problems to try it on are:
1. Particle in a box
2. Simple harmonic oscillator

If you solve these problems, try the harder case of finding the ground state of the hydrogen atom. Alternatively, explore tutorials for the package `netket` to get a sense of research level problems that are being addresses with neural quantum states.
